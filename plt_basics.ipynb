```python
import matplotlib.pyplot as plt
```

```python
# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Plotting the data
plt.plot(x, y)

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Plot')

#Displaying the plot
plt.show()  
```

```python
# Plotting the data
plt.plot(x, y)

# Adding a title
plt.title('Sample Line Plot')

# Adding labels to the axes
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Displaying the plot
plt.show()
```

```python
# Random data points
x = [5, 2, 9, 4, 7]
y = [10, 5, 8, 4, 2]

# Creating the scatter plot
plt.scatter(x, y)

# Adding title and labels
plt.title('Sample Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Displaying the plot
plt.show()
```

```python
# Categories and values
categories = ['Red', 'Blue', 'Green', 'Purple']
values = [10, 15, 7, 5]

# Creating the bar chart
plt.bar(categories, values)

# Adding title and labels
plt.title('Simple Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')

# Displaying the plot
plt.show()
```

```python
# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Plotting with a specific color and marker
plt.plot(x, y, marker='s', color='green')  # 's' stands for square marker

# Adding title and labels
plt.title('Colorful Plot with Square Markers')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Displaying the plot
plt.show()
```
