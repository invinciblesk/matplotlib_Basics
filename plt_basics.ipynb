```python
import matplotlib.pyplot as plt
```

```python
# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Plotting the data
plt.plot(x, y)

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Plot')

#Displaying the plot
plt.show()  
```

```python
# Plotting the data
plt.plot(x, y)

# Adding a title
plt.title('Sample Line Plot')

# Adding labels to the axes
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Displaying the plot
plt.show()
```
